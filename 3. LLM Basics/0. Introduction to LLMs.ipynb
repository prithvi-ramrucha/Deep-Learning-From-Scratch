LLM stands for *L*arge *L*anguage *M*odel and is an umbrella term that describes a family of transformer-based neural networks that were orginally intended for natural language processing (NLP) tasks. These tasks included, chatting, translation, summarisation, sentiment analysis, etc. However, due to the large number of trainable parameters (at least, so we believe), billions or even trillions, LLMs have demonstrated reasoning-like capabilities far exceeding their original purpose. We know what LLMs are capable of, but what exactly is an LLM? At this stage, we can imagine an LLM as a _"next word predictor"_. Given a sequence $S$ of tokens, an LLM outputs a vector $\mathbf{z}$ containing **logits** associated with each possible output token. $$

$$

### 0.1 Emergence & Prompting

In the early era of LLMs, the main paradigm in development was that more parameters means better model performance. This scaling law arose from the emergent properties displayed by the first LLM models. GPT-3, often considered the first true LLM, displayed reasoning-like skills such as few-shot/zero-shot learning, basic arithmetic, chain-of-thought logic, writing code, etc. These capabilties extended outside of the typical NLP tasks it was orginally trained for. Moreover, GPT-3 could exhibit all these properties without further fine-tuning; it was a jack-of-all-trades. Instead of fine-tuning, the behaviour is brought out via **prompting**. Some common types of prompting,

| Prompt Type | Description | Example |
|-------------|-------------|---------|
| Zero-shot learning | The model is given only an instruction with no examples of input-output pairs. It must rely entirely on pre-trained knowledge to respond. | "Translate the following sentence into French: 'I am learning machine learning.'" |
| Single-shot learning | The prompt includes exactly one example of an input-output pair to guide the model’s behavior. | "English: 'I am tired.' → French: 'Je suis fatigué.'\nEnglish: 'I am learning machine learning.' → ?" |
| Few-shot learning | The prompt includes multiple examples (typically 2–10) of input-output pairs to demonstrate the task pattern. | "English: 'Good morning.' → French: 'Bonjour.'\nEnglish: 'How are you?' → French: 'Comment ça va ?'\nEnglish: 'I am learning machine learning.' → ?" |
| Role prompting | The model is instructed to adopt a specific persona or role to shape its style, tone, or reasoning behavior. | "You are a strict mathematics professor. Explain what a derivative is." |
| Delimiting context | Input context is clearly separated from instructions using delimiters (e.g., quotes, XML tags, triple backticks) to reduce ambiguity. | "Summarise the text between triple quotes:\n\"\"\"LLMs are transformer-based models trained on large datasets.\"\"\"" |
| Chain-of-thought prompting | The model is encouraged to explicitly show intermediate reasoning steps before producing a final answer. | "Solve step by step: If a car travels 60 km in 2 hours, what is its speed?" |

Increasing the dataset size and compute power assigned to training can also lead to emergent behaviour. Regarding the number of trainable parameters, LLMs typically displayed strong emergent behaviour after a few billion parameters. After emergence, scaling laws predict incremental, predictable improvements rather than dramatic leaps when increasing the number of parameters in a model, dataset size or computing power assigned to training. One most always consider the trade-off between these factors. Analysis demonstrates that the best results come from balancing model and dataset sizes.

### 0.3 Quantifying LLM Sizes

LLMs contain billions or even trillions of parameters. We can calculate how much storage is needed to store the parameters of an LLM by knowing its number of trainable parameters and their floating-point precision. Take the legacy model GPT-3 for example, it has `175,000,000,000` trainable parameters stored as `32-bit` floats. Given that a `4` bytes are needed to store each `32-bit` float, the storage required for GPT-3 is `4 x 175,000,000 = 700,000,000,000` bytes or `700 GB`. While never public confirmed, state-of-the-art models such as GPT-5 are likely to contain trillions of parameters and require terabytes of storage. A common way to reduce LLM model sizes while minmising a drop in performance is to use represent the model parameters in lower floating point precision such as `8-bit` and `4-bit`. This process is known as quantisation.

### 0.4 Hyperparameter Sensitivity

Due to the large size of LLMs, they are essentially sensitive to changes in the hyperparameters. For example, if a shift in the learning rate $\eta$ causes just 1% of weight updates in a 1 billion parameter LLM to become unstable, there will be millions of unstable weight updates. Moreovoer, the robustness of LLMs to prompt injection attacks is also heavily influenced by its hyperparameters. These hyperparameters include tokenisation methods, learning rate, scheduling strategies, dropout rate and weight decay.
