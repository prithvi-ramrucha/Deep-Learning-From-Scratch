# **1. Embeddings**

Tokenisation is splitting a sentence into smaller substrings called tokens. For example, 

> **The quick brown fox jumps over the lazy dog.**

is split into the tokens [...].

### 1.2 Byte Pair Encoding

BPE was orginally intended to compress text, but was applied in tokenisation for early GPT models by OpenAI. In BPE, we start off with a large corpus (large collection of text). Using this corpus, we build a vocabulary by taking all characters and symbols used to write the words. In pratical settings, this vocabulary will contain all ASCII characters and many other symbols.
