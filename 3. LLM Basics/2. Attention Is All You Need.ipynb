# **2. Attention Is All You Need**

### 2.1 Motivation

Attention layers are a core component of LLMs. Given a series of token embeddings, attention layers update each embedding based on the other embeddings in the sequence using weights. These weights are determined by a similarity measure between embeddings (which pairs of embeddings are most related to each other). Very loosely speaking, attention layers enable token embeddings to encode rich contextual information. To build intuition, consider the sentence:

> **The quick brown fox jumps over the lazy dog.**

We assume that each token roughly corresponds to each individual word. Intuitively, we may propose the idea that an attention layer updates the embedding tokens corresponding to the nouns _"fox"_ and _"dog"_ based on the adjectives _"quick"_ and _"lazy"_ respectively. In this way, the representation of each token becomes dependent on other semantically related tokens in the sequence. Another attention layer could update the nouns based on the adverb "jump". While in practice, layers may not correspond to these exact linguistic roles, this example illustrates the key idea: each token representation is iteratively refined by incorporating information from other embeddings in the sequence.
