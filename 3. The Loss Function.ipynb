# 3.1 Categorical Cross-Entropy

A **loss function** provides a metric for how wrong a neural network model is in its prediction. That is, it numerically quantifies how different or "wrong" our model's prediction is to the target output. In cases, where we are training our neural network to solve a _classification_ problem, **categorical cross-entropy** is a common choice of a loss function. This is because it deals with probability distrubutions. In full, it is defined as,

$$ L_i = -\sum_{j}Y_{ij} log(\hat{Y_{ij}}) $$
