# 0. Introduction to PINNs

PINNs are **p**hysics-**i**nformed **n**eural **n**etworks. In the original paper by Raissi et al. (2019), they were introduced as MLPs with heavily modified loss functions that included physical constraints derived from known partial differential equations. The dynamics of many physical phenomena in fluids, quantum mechanics, mechanical systems, etc. are governed by partial differential equations. These dynamics can be incorporated into the loss function $\mathcal{L}$ of by adding residue terms that must be minimised. In doing so, one can construct an objective function $J$ consisting of a typical ML loss $\mathcal{L}$ function (MSE for example) and the residual terms constructed from PDEs which describe the physical system. Let us consider the quintessential example, that of the flow of an incompressible fluid. Ideally, such a fluid follows the Navier-stokes equations and continuity equation,

$$
\frac{\partial \mathbf{u}}{\partial t} + (\mathbf{u} \cdot \nabla)\mathbf{u} = - \nabla p + \frac{1}{\text{Re} } \nabla^2 \mathbf{u}
$$

$$
\nabla \cdot \mathbf{u} = 0
$$
